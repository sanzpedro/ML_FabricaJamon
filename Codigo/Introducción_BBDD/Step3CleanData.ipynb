```python
import pandas as pd
import numpy as np
```

```python
# Defino la función de consulta:

def ConsultaBBDD(a):
    import mysql.connector
    import pandas as pd
    from sqlalchemy import create_engine

    cadena_conexion = 'mysql+pymysql://root@localhost:3306/factoria_jamones'
    conexion = create_engine(cadena_conexion)
    query = pd.read_sql_query(str(a), con=conexion)
    return (query)
```

```python
QueryTablaPartida = "select id_partida, no_partida, añada, desc_raza, desc_alimentacion, acido_oleico, acido_palmitico, acido_estearico, acido_linoleico, peso_varios, peso_vientres, peso_cerdos, peso_neto_canal, nacimiento, desc_matadero, no_animales, desc_tipo_pesada_entrada, fecha_entrada from partida, alimentacion, raza, matadero, tipo_pesada_entrada where partida.id_alimentacion = alimentacion.id_alimentacion and partida.id_raza = raza.id_raza and partida.id_matadero = matadero.id_matadero and partida.id_tipo_pesada_entrada = tipo_pesada_entrada.id_tipo_pesada_entrada "
```

```python
df_partida = ConsultaBBDD(QueryTablaPartida)
df_partida
```

```python
QueryTablaPieza = "select id_pieza, desc_tipo_pieza, id_partida, no_serie, peso_entrada, peso_salida, fecha_salida, desc_tipo_pesada_salida, desc_estado_salida from pieza, tipo_pieza, tipo_pesada_salida, estado_salida where pieza.id_tipo_pieza = tipo_pieza.id_tipo_pieza and pieza.id_tipo_pesada_salida = tipo_pesada_salida.id_tipo_pesada_salida and pieza.id_estado_salida = estado_salida.id_estado_salida"
```

```python
df_pieza = ConsultaBBDD(QueryTablaPieza)
df_pieza
```

```python
df_completo = df_pieza.join(df_partida, on=["id_partida"], how="inner",lsuffix="str")
for cols in df_completo:
    df_completo[cols] = df_completo[cols].replace(np.nan, "NULL")
df_completo
```

```python

df_v1 = df_completo.copy()

# df_v1[df_v1["desc_tipo_pieza"]=="Jamón"].groupby("Añada").agg("count")
# df_v1[df_v1["desc_tipo_pieza"]=="Paleta"].groupby("Añada").agg("mean")



L2pieza = []
L2raza = []
L2alimentacion = []
L2mediaE = []
L2mediaS = []
L2desvE = []
L2desvS = []
L2countE = []
L2countS = []

diccPieza = {}
diccPieza["Jamón"] = [[8, 20],[5, 16.4]]
diccPieza["Paleta"] = [[6, 15],[2.5, 12]]
# diccPieza["Jamón"] = [[8, 18],[5, 15]]
# diccPieza["Paleta"] = [[6, 12],[3, 9]]


for pieza in df_v1["desc_tipo_pieza"].unique():
    for raza in df_v1["desc_raza"].unique():
        for alimentacion in df_v1["desc_alimentacion"].unique():
            mediaE = df_v1["peso_entrada"][(df_v1['desc_alimentacion']==alimentacion)&(df_v1['desc_raza']==raza)&(df_v1['desc_tipo_pieza']==pieza)&(df_v1['peso_entrada'].astype("float").between(diccPieza[pieza][0][0],diccPieza[pieza][0][1]))].mean()
            mediaS = df_v1["peso_salida"][(df_v1['desc_alimentacion']==alimentacion)&(df_v1['desc_raza']==raza)&(df_v1['desc_tipo_pieza']==pieza)&(df_v1['peso_salida'].astype("float").between(diccPieza[pieza][1][0],diccPieza[pieza][1][1]))].mean()
            desvE = df_v1["peso_entrada"][(df_v1['desc_alimentacion']==alimentacion)&(df_v1['desc_raza']==raza)&(df_v1['desc_tipo_pieza']==pieza)&(df_v1['peso_entrada'].astype("float").between(diccPieza[pieza][0][0],diccPieza[pieza][0][1]))].std()
            desvS = df_v1["peso_salida"][(df_v1['desc_alimentacion']==alimentacion)&(df_v1['desc_raza']==raza)&(df_v1['desc_tipo_pieza']==pieza)&(df_v1['peso_salida'].astype("float").between(diccPieza[pieza][1][0],diccPieza[pieza][1][1]))].std()
            countE = df_v1["peso_entrada"][(df_v1['desc_alimentacion']==alimentacion)&(df_v1['desc_raza']==raza)&(df_v1['desc_tipo_pieza']==pieza)&(df_v1['peso_entrada'].astype("float").between(diccPieza[pieza][0][0],diccPieza[pieza][0][1]))].count()
            countS = df_v1["peso_salida"][(df_v1['desc_alimentacion']==alimentacion)&(df_v1['desc_raza']==raza)&(df_v1['desc_tipo_pieza']==pieza)&(df_v1['peso_salida'].astype("float").between(diccPieza[pieza][1][0],diccPieza[pieza][1][1]))].count()
            
            
            L2pieza.append(pieza)
            L2raza.append(raza)
            L2alimentacion.append(alimentacion)
            L2mediaE.append(mediaE)
            L2mediaS.append(mediaS)
            L2desvE.append(desvE)
            L2desvS.append(desvS)
            L2countE.append(countE)
            L2countS.append(countS)
            
            

            import seaborn
            import matplotlib.pyplot as plt
            fig= plt.subplots(figsize=(20,13), dpi= 80)
            seaborn.distplot(df_v1["peso_entrada"][(df_v1['desc_alimentacion']==alimentacion)&(df_v1['desc_raza']==raza)&(df_v1['desc_tipo_pieza']==pieza)&(df_v1['peso_entrada'].astype("float").between(diccPieza[pieza][0][0],diccPieza[pieza][0][1]))], label="Pesos %s %s %s Entrada" %(pieza, raza, alimentacion))
            seaborn.distplot(df_v1["peso_salida"][(df_v1['desc_alimentacion']==alimentacion)&(df_v1['desc_raza']==raza)&(df_v1['desc_tipo_pieza']==pieza)&(df_v1['peso_salida'].astype("float").between(diccPieza[pieza][1][0],diccPieza[pieza][1][1]))], label="Pesos %s %s %s Salida" %(pieza, raza, alimentacion))
            plt.legend()
           
dicc2 = {'desc_pieza': L2pieza, 'desc_raza': L2raza, 'desc_alimentacion': L2alimentacion,'mediaE': L2mediaE,'mediaS': L2mediaS,'desvE': L2desvE,'desvS': L2desvS,'countE': L2countE,'countS': L2countS}
            
df_estadisticas2 = pd.DataFrame(data=dicc2)
df_estadisticas2
```
